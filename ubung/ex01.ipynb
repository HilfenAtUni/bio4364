```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('ex01.csv', sep=',')
name = data.columns[1:]
studiengang = data.values[0,1:]
semester = data.values[1,1:]
lang = data.values[3:10, 0]
lang_data = data.values[3:10, 1:]
tools = data.values[13:23, 0]
tools_data = data.values[13:23, 1:]
```

```python
fig = plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
plt.plot(semester, '.');
plt.title('which semester of all the student')
plt.grid(True)
plt.subplot(1,2,2)
plt.plot(pd.value_counts(semester), '.')
plt.title('the number of the student by semester')
plt.grid(True)

fig = plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(studiengang, '.');
plt.title('which studiengang of all the student')
plt.grid(True)
plt.subplot(1,2,2)
plt.plot(pd.value_counts(studiengang), '.')
plt.title("the number of student by studiengang")
plt.xticks(rotation=270)
plt.grid(True)
```

```python
# different language
lang_data[lang_data == '1'] = 1
lang_data[lang_data == '2'] = 2
lang_data[lang_data == '3'] = 3
lang_data[lang_data == '4'] = 4
lang_data[lang_data == '5'] = 5
fig = plt.figure(figsize=(16,10))
for idx, l in enumerate(lang_data):
    plt.subplot(2, 4, idx+1)
    plt.plot(l, '.')
    title = lang[idx]
    plt.title("[{}]\nmean:{:.3f}, std:{:.3f}, var:{:.3f}".format(
                            title, l.mean(), l.std(), l.var()))
    plt.xlim(0,29)
    plt.ylim(.8, 5.2)
    plt.grid(True)
```

- Conclusion

    From the tables, Java, Python are the best two chooices. What's more, Perl will not be accepted, and PHP is also not a good choice.

```python
# different tools
tools_data[tools_data == '1'] = 1
tools_data[tools_data == '0'] = 0

fig = plt.figure(figsize=(16,12))
for idx, t in enumerate(tools_data):
    plt.subplot(3, 4, idx+1);
    plt.plot(t, '.');
    plt.ylim(-.2, 1.2)
    plt.title("[{}] 'Yes': {}, 'No':{}".format(
            tools[idx], np.size(t[(t==1)==True]), np.size(t[(t==0)==True])));
    plt.grid(True)
```

- Conclusion

    From this part, it shwos that matplotlib is most popular. What's is the most useful information is that, none of us will choose Mauve.
